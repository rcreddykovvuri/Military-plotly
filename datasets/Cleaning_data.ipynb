```python

print(df)
```

```python
df
```

```python
#replacing comma(,) with none and changing the datatype to int
for x in comma:
    df[x] = df[x].str.replace(',','')
    df[x] = df[x].astype(int)
```

```python
print(df)
```

```python
#changing column datatype to int
df['aircraft_strength'] = df['aircraft_strength'].astype(int)
```

```python
#changing column datatype to int
df['fighter/interceptor_strength'] = df['fighter/interceptor_strength'].astype(int)
df['attack_aircraft_strength'] = df['attack_aircraft_strength'].astype(int)
df['transport_aircraft_fleet_strength'] = df['transport_aircraft_fleet_strength'].astype(int)
df['trainer_aircraft_fleet'] = df['trainer_aircraft_fleet'].astype(int)
```

```python
#changing column datatype to int
df['helicopter_fleet_strength'] = df['helicopter_fleet_strength'].astype(int)
df['attack_helicopter_fleet_strength'] = df['attack_helicopter_fleet_strength'].astype(int)
df['tank_strength'] = df['tank_strength'].astype(int)
df['afv/apc_strength'] = df['afv/apc_strength'].astype(int)
```

```python
#changing column datatype to int
df['self-propelled_artillery_strength'] = df['self-propelled_artillery_strength'].astype(int)
df['towed_artillery_strength'] = df['towed_artillery_strength'].astype(int)
df['rocket_projector_strength'] = df['rocket_projector_strength'].astype(int)
df['navy_fleet_strengths'] = df['navy_fleet_strengths'].astype(int)
df['aircraft_carrier_fleet_strength'] = df['aircraft_carrier_fleet_strength'].astype(int)
df['submarine_fleet_strength'] = df['submarine_fleet_strength'].astype(int)
df['destroyer_fleet_strength'] = df['destroyer_fleet_strength'].astype(int)
df['frigate_fleet_strength'] = df['frigate_fleet_strength'].astype(int)
```

```python
#changing column datatype to int
df['airport_totals'] = df['airport_totals'].astype(int)
```

```python
#importing country_converter to get iso values
import country_converter as coco
```

```python
#getting iso values for countries
out_format = "ISO3"
df['iso'] = coco.convert(names=df['country'], to=out_format, not_found=None)
```

```python
print(df['iso'])
```

```python
#replacing strenth with none
df.columns = df.columns.str.replace('_strength','')
```

```python
import pycountry_convert as pc

def country_to_continent(country_name):
    country_alpha2 = pc.country_name_to_country_alpha2(country_name)
    country_continent_code = pc.country_alpha2_to_continent_code(country_alpha2)
    country_continent_name = pc.convert_continent_code_to_continent_name(country_continent_code)
    return country_continent_name
```

```python
#get continents for countries
country = list(df['country'])
continent = []
for i in country:
    continent.append(country_to_continent(i))
```

```python
df['continent'] = continent
```

```python
#saving the file to csv
df.to_csv('cleaned.csv',index=False)
```
